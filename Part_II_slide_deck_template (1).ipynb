## Investigation Overview

> The goal is to see those features that affects the total biking duration of each participant in the dataset.

## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area collected through January 2018 to December 2018


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
#create a new column for ages of the users using their birth year up till 2018
df['member_age']=2018-df.member_birth_year
```

```python
#create a new column for duration of biking in hours
df['duration_hr']=df.duration_sec/3600
```

## (Visualization 1)

> The duration of biking is spread over a wide a range of time (in seconds) so a log scale was employed to make a convenient plot. it is observed that a large portion of the participants spend about 1,000s biking.


```python
log_binsize = 0.025
bins = 10 ** np.arange(2.4, np.log10(df['duration_sec'].max())+log_binsize, log_binsize)

#plt.figure(figsize=[10, 8])
plt.hist(data = df, x = 'duration_sec', bins = bins)
plt.title('Biking Duration')
plt.xscale('log')
plt.xlim([250, 2e4])
plt.xlabel('Time spent(in seconds)')
plt.ylabel('Number Of people')
plt.show()
```

## (Visualization 2)

> A large population of the people in the dataset are subscriber and majorly below the age of 80 years. A large proportion rode less than 5 hours according to the plot

```python
#Distribution of member age

#plt.figure(figsize=(12,8))
ax=sns.scatterplot(df.duration_hr,df.member_age, hue=df.user_type)
plt.title('Duration of biking by members age')
plt.xlabel('Duration of biking')
plt.ylabel('Members age')
plt.show()
```

## (Visualization 3)

> The plot shows that the larger population of the people in the dataset are males and majority of them spent more time biking than others

```python
#plt.figure(figsize=(12,8))
ax=sns.scatterplot(df.duration_hr,df.member_age, hue=df.member_gender)
plt.title('Duration of biking by members age')
plt.xlabel('Duration of biking')
plt.ylabel('Members age')
plt.show()
```
